```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

im_dir = r'E:/Semester 4/EN2550_FundamentalsOfImageProcessing/Fun-Time/Exercises/images/'
```

```python
f=cv.imread(im_dir+'spider.png')
assert f is not None

fig, ax =plt.subplots() 
ax.imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB))
plt.show()
```

```python
#Linear Transformation
t=np.arange(255,-1,-1,dtype=np.uint8)
g=cv.LUT(f,t)
# g=t[f]

fig,ax=plt.subplots()
ax.plot(t)
```

```python
fig, ax =plt.subplots() 
ax.imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
plt.show()
```

```python
#Gamma Correction
gamma_list=[0.2,0.8,1.2,2]
fig, ax =plt.subplots(2,2,figsize=(16,16)) 

for (i,gamma) in enumerate(gamma_list):
    t=np.array([((p/255)**gamma)*255 for p in range (0,256)]).astype(np.uint8)
    g=cv.LUT(f,t)
    
    ax[i//2][i%2].imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
    ax[i//2][i%2].set_title("gamma = %f"%gamma)

plt.show()

```

```python
t1=np.linspace(0,100,50)
t2=np.linspace(100,255,150)
t3=np.linspace(255,255,56)

t=np.concatenate((t1,t2,t3),axis=0).astype(np.uint8)
fig,ax=plt.subplots()
ax.plot(t)
```

```python
g=cv.LUT(f,t)
fig, ax =plt.subplots() 
ax.imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
plt.show()
```

```python
cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',f)
cv.waitKey(0)
cv.imshow('Image',g)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
f1=cv.imread(im_dir+'shells.tif',cv . IMREAD_GRAYSCALE)
assert f1 is not None

fig, ax =plt.subplots() 
ax.imshow(f1,cmap='gray',vmin=0,vmax=255)
plt.show()
```

```python
hist_f=cv.calcHist([f1],[0],None,[256],[0,256])
f2=cv.equalizeHist(f1)
hist_g=cv.calcHist([f2],[0],None,[256],[0,256])


fig,ax=plt.subplots(2,1,figsize=(8,8))
ax[0].plot(hist_f)
ax[0].set_title("Histogram of the Original Image")
ax[1].plot(hist_g)
ax[1].set_title("Histogram of the Equalized Image")
```

```python
#Displaying Equalized Image
fig, ax =plt.subplots() 
ax.imshow(f2,cmap='gray',vmin=0,vmax=255)
plt.show()
```

```python
f4=cv.imread(im_dir+'zion_pass.jpg')
assert f4 is not None

fig, ax =plt.subplots() 
ax.imshow(cv.cvtColor(f4, cv.COLOR_BGR2RGB))
plt.show()
```

```python
#Changing the saturation of the image
hlsImg = cv.cvtColor(f4, cv.COLOR_BGR2HLS)

hlsImg[:,:,2]=np.clip(hlsImg[:,:,2]+50,0,255)

lsImg = cv.cvtColor(hlsImg, cv.COLOR_HLS2BGR)


fig, ax =plt.subplots() 
ax.imshow(cv.cvtColor(lsImg, cv.COLOR_BGR2RGB))
plt.show()
```

```python
#Changing the hue
hlsImg = cv.cvtColor(f4, cv.COLOR_BGR2HLS)

t=np.arange(255,-1,-1,dtype=np.uint8)
hlsImg[:,:,0]=cv.LUT(hlsImg[:,:,0],t)

lsImg = cv.cvtColor(hlsImg, cv.COLOR_HLS2BGR)

fig, ax =plt.subplots() 
ax.imshow(cv.cvtColor(lsImg, cv.COLOR_BGR2RGB))
plt.show()
```
